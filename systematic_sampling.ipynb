```python
# Import packgae numpy
import numpy as np
# Import packgae pandas
import pandas as pd
 
# Buat data frame dengan membaca dataset
df = pd.read_csv("../Dataset/students.csv")

# Function Systematic Sampling
def systematic_sampling(df, step):
    indexes = np.arange(0, len(df), step = step)
    systematic_sample = df.iloc[indexes]
    return systematic_sample

# Dapatkan sampel sistematis dan simpan dalam variabel systematic_sample
systematic_sample = systematic_sampling(df, 100)
 
# Tampilkan data frame
display(systematic_sample)
 
# Hitung rata rata math score dari data sample
systematic_data = round(systematic_sample['math score'].mean())

# Tampilkan hasil perhitungan rata rata
print("Average Math Score :", systematic_data)
```
